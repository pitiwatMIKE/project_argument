```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))


models = ['CRF','LSTM','BiLSTM','BiLSTM-CRF', 'Wangchanberta']
f1_claim = [0.37, 0.15, 0.17, 0.23, 0.51]
f1_premise = [0.32, 0.16, 0.17, 0.13, 0.39]

# Set position of bar on X axis
br1 = np.arange(len(f1_claim))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, f1_claim, width = barWidth, label ='Claim')
plt.bar(br2, f1_premise, width = barWidth, label ='Premise')


plt.xlabel('Model Name', fontweight ='bold', fontsize = 13)
plt.ylabel('F1 Score', fontweight ='bold', fontsize = 13)
plt.xticks([r + barWidth for r in range(len(f1_claim))],models)
plt.legend()
plt.savefig('sentence.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.12
fig = plt.subplots(figsize =(20, 12))


models = ['CRF','LSTM','BiLSTM','BiLSTM-CRF', 'Wangchanberta']
B_claim = [0.66, 0.32, 0.76, 0.41, 0.75]
B_premise = [0.56, 0.59, 0.58, 0.34, 0.64]
I_claim = [0.50, 0.54, 0.59, 0.32, 0.65]
I_premise = [0.72, 0.69, 0.67, 0.48, 0.72]

# Set position of bar on X axis
br1 = np.arange(len(f1_claim))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

# Make the plot
plt.bar(br1, B_claim, width = barWidth, label ='B-claim')
plt.bar(br2, B_premise, width = barWidth, label ='B-premise')
plt.bar(br3, I_claim, width = barWidth, label ='I-claim')
plt.bar(br4, I_premise, width = barWidth, label ='I-premise')


plt.xlabel('Model Name', fontweight ='bold', fontsize = 20)
plt.ylabel('F1 Score', fontweight ='bold', fontsize = 20)
plt.xticks([r + barWidth for r in range(len(f1_claim))],models, fontsize=18)
plt.legend(prop={'size': 15})
plt.savefig('word.png')
plt.show()
 
```
